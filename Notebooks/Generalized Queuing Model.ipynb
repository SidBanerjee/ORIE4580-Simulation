```python
# Import the usual libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# General class for the queues themselves
class Queue(object):
    
    name = ""
    waitingentities = []
    
    # Initialize a Queue
    def __init__(self, name, waiting):
        self.name = name
        self.waitingentities = waiting
        
    # Print a Queue (just its name and the entire list of waiting Entities)
    def __repr__(self):
        return self.name + ": " + str(self.waitingentities)
    
    # A new Entity is created and arrives at this Queue
    def ArriveatQueue(self, time):
        newarrival = Entity(time)
        self.waitingentities.append(newarrival)
        
    # Exit the Queue (and the system)
    def LeaveQueue(self):
        newentities = []
        for e in range(1, np.size(self.waitingentities)):
            newentities.append(self.waitingentities[e])
        self.waitingentities = newentities
        
    # Send the first Entity in line to another Queue
    def SendTo(self, destination):
        destination.waitingentities.append(self.waitingentities[0])
        newentities = []
        for e in range(1, np.size(self.waitingentities)):
            newentities.append(self.waitingentities[e])
        self.waitingentities = newentities
```

```python
# General class for the entities that move through the Queues
class Entity(object):
    
    starttime = 0
    
    # Initialize an Entity
    def __init__(self, start):
        self.starttime = start
        
    # Print an Entity (just arrival time)
    def __repr__(self):
        return str(self.starttime)
```

```python
# Create queues (you will need to create a Queue for every station in the clinic)
line1 = Queue("Line 1", [])
# Line 1 is the last stop in our system
line2 = Queue("Line 2", [])
# Line 2 is an intermediary step, sends people to Line 1
line3 = Queue("Line 3", [])
# Line 3 can send people to either Line 1 or Line 2

# Store data for plot (you can store whatever data you like in lists like this)
times = []
q1len = []
q2len = []
q3len = []
totalwaits = []

# General discrete-event loop
# Initialize times
currenttime = 0
nextarrival1 = np.random.exponential(4)
nextdeparture1 = 1e6
nextarrival2 = np.random.exponential(4)
nextdeparture2 = 1e6
nextarrival3 = np.random.exponential(4)
nextdeparture3 = 1e6
# Make an array of all the "next times"
nexteventtimes = np.array([nextarrival1, nextdeparture1, nextarrival2, nextdeparture2, nextarrival3, nextdeparture3])
while currenttime < 100:
    #Find which "next time" is next, set time to that
    eventtype = np.argmin(nexteventtimes)
    currenttime = np.amin(nexteventtimes)
    print("Current time: ", currenttime)
    
    # Do the events corresponding to whatever the "soonest" next time is
    if eventtype == 0:
        print("Arrive at Line 1")
        nexteventtimes[0] += np.random.exponential(4)
        line1.ArriveatQueue(currenttime)
        if nexteventtimes[1] == 1e6:
            nexteventtimes[1] = currenttime + np.random.exponential(1)
        
    if eventtype == 1:
        print("Leave Line 1")
        totalwaits.append(currenttime - line1.waitingentities[0].starttime)
        line1.LeaveQueue()
        if np.size(line1.waitingentities) > 0:
            nexteventtimes[1] += np.random.exponential(1)
        else:
            nexteventtimes[1] = 1e6
        
    if eventtype == 2:
        print("Arrive at Line 2")
        nexteventtimes[2] += np.random.exponential(4)
        line2.ArriveatQueue(currenttime)
        if nexteventtimes[3] == 1e6:
            nexteventtimes[3] = currenttime + np.random.exponential(1)
        
    if eventtype == 3:
        print("Move from 2 to 1")
        line2.SendTo(line1)
        if nexteventtimes[1] == 1e6:
            nexteventtimes[1] = currenttime + np.random.exponential(1)
        if np.size(line2.waitingentities) > 0:
            nexteventtimes[3] += np.random.exponential(1)
        else:
            nexteventtimes[3] = 1e6
            
    if eventtype == 4:
        print("Arrive at Line 3")
        nexteventtimes[4] += np.random.exponential(4)
        line3.ArriveatQueue(currenttime)
        if nexteventtimes[5] == 1e6:
            nexteventtimes[5] = currenttime + np.random.exponential(1)
    
    if eventtype == 5:
        print("Depart Line 3")
        decidewhere = np.random.uniform()
        if decidewhere > .5:
            print("Go to Line 2")
            line3.SendTo(line2)
            if nexteventtimes[3] == 1e6:
                nexteventtimes[3] = currenttime + np.random.exponential(1)
        else:
            print("Go to line 1")
            line3.SendTo(line1)
            if nexteventtimes[1] == 1e6:
                nexteventtimes[1] = currenttime + np.random.exponential(1)
        if np.size(line3.waitingentities) > 0:
            nexteventtimes[5] += np.random.exponential(1)
        else:
            nexteventtimes[5] = 1e6
    
    print(line1)
    print(line2)
    print(line3)
        
    # Record data
    times.append(currenttime)
    q1len.append(np.size(line1.waitingentities))
    q2len.append(np.size(line2.waitingentities))
    q3len.append(np.size(line3.waitingentities))
```

```python
# Make an informative plot of queue lengths over time
plt.figure(figsize=(12,8))
plt.step(times[:20], q1len[:20], label="Line 1", where='post')
plt.step(times[:20], q2len[:20], label="Line 2", where='post')
plt.step(times[:20], q3len[:20], label="Line 3", where='post')
plt.title("3 Queue System")
plt.ylabel("People Waiting")
plt.xlabel("Time")
plt.legend(title="Queue Length")
plt.show()
```

```python
# Make a histogram of total times in the system
plt.figure(figsize=(12,8))
plt.hist(totalwaits)
plt.title("Total Time in System")
plt.ylabel("Number of People")
plt.xlabel("Total Time in System")
plt.show()
```

```python
# Find the mean time in system
print("Average Time in System", np.mean(totalwaits))
```
