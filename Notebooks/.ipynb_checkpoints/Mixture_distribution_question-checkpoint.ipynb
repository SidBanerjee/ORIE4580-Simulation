```python
# Preamble to make sure we have all packages we need

import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
%matplotlib inline

# Choosing a colormap for the plot colors
cmap=plt.get_cmap('Set3')

import seaborn as sns
sns.set_style("white")
sns.set_context("talk",font_scale=1, rc={"lines.linewidth": 2.5})
```

```python
def emp_cdf_discrete(x):
    n_max = np.max(x)
    x_sort = np.arange(n_max+1)
    cdf = np.cumsum(np.bincount(x))
    cdf = cdf/np.max(cdf)
    return x_sort, cdf
```

## Modeling length of stay at a hotel

An hotel owner wants to understand the distribution of the length of stay of her past guests. She believes that the distribution is a mixture distribution, where a fraction $\alpha$ of guests stay for one night, while the length-of-stay of the remaining $(1-\alpha)$ fraction of guests can be modeled as 1+T, where T is a Poisson distribution with parameter $\lambda$. Given past data, we want to estimate $\alpha$ and $\lambda$.

```python
N = 600
alpha = 0.4

lam = 5

n_alpha = np.random.binomial(N,alpha)
X = np.concatenate([1+np.random.poisson(lam = lam,size=N-n_alpha), 
                    np.array(n_alpha*[1])])

sns.distplot(X,kde=False,norm_hist=True)
plt.legend(loc=1)
plt.show()
```

```python
x,cdf = emp_cdf_discrete(X)
plt.step(x,cdf,where='post',linestyle=':')

y = np.append(0,(1-alpha)*sc.stats.poisson.cdf(x[:-1],lam))
y[1:]=y[1:]+alpha

#plt.step(x, y, 'b--',where='post',linewidth=1,color='m',label='Original cdf')

lam_mle = np.mean(X)
z = sc.stats.poisson.cdf(x,lam_mle)
plt.step(x, z, 'b--',where='post',linewidth=1,color='m',label='Original cdf')
plt.show()
```

### MOM estimator for the mixture

We need to find two parameters: $\alpha$ and $\lambda$. Note that we have:
$$m_1 = \alpha + (1-\alpha)(1+\lambda) = 1 + (1-\alpha)\lambda$$
$$m_2 = \alpha + (1-\alpha)(1+3\lambda+\lambda^2) = 1+(1-\alpha)\lambda(3+\lambda)$$
Simplifying we get: $$\lambda = \frac{m_2-1}{m_1-1}-3 \quad , \quad\alpha = 1-\frac{(m_1-1)^2}{2+m_2-m_1}$$
