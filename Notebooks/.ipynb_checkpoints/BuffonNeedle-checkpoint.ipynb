```python
def buffon(N,length,dist,seed=-1):
    # N = number of throws 
    # length = length of match
    # dist = distance between lines
    # seed = seed for random number generator
    
    if seed==-1:
        xcent = np.random.uniform(0,dist/2.0,N)
        theta = np.random.uniform(0,math.pi/2,N)
    else:
        np.random.seed(seed)
        xcent = np.random.uniform(0,dist/2.0,N)
        theta = np.random.uniform(0,math.pi/2,N)

    xtip_l  = xcent - (length/2.0)*np.cos(theta)
    data = (xtip_l < 0)      
    return data
    
    
#    for i in range(n):
#        hits = 0
#        xcent = np.random.uniform(0,b/2.0)
#        theta = np.random.uniform(0,math.pi/2)
#        xtips  = (xcent - (a/2.0)*math.cos(theta),xcent + (a/2.0)*math.cos(theta))
#        if xtips[0] < 0 :
#            hits = 1
#        else:
#            hits = 0            
#        data.append(hits)
#    return data
```
